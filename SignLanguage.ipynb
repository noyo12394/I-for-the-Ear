```python
from ultralytics import YOLO

# Load the YOLOv8 nano model for faster training
model = YOLO('/Users/nchatterjee/Downloads/yolov8n.pt')

# Train the model with optimized settings for faster completion
model.train(
    data='/Users/nchatterjee/Downloads/American Sign Language Letters/data.yaml',  # Path to dataset
    epochs=10,               # Reduce epochs to 10 for faster training
    imgsz=416,               # Reduce image size to 416x416 for speed
    batch=16,                # Increase batch size if memory allows, adjust as needed
    workers=4,               # Use multiple workers to speed up data loading
    project='/Users/nchatterjee/Downloads/yolo_outputs',  # Output directory
    name='sign_language_detection',  # Name of the training run
    optimizer='Adam'        # Use Adam optimizer for faster convergence
)

# Evaluate the trained model
metrics = model.val()
print(metrics)

```

```python
from ultralytics import YOLO
import cv2

# Load your trained model
model = YOLO('/Users/nchatterjee/Downloads/yolo_outputs/sign_language_detection3/weights/best.pt')

# Capture video from the camera
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection
    results = model.predict(source=frame, save=False)

    # Display results with class labels and confidence scores
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = f"{model.names[int(box.cls[0])]}: {box.conf[0]:.2f}"
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Show the frame
    cv2.imshow('Sign Language Detection', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
